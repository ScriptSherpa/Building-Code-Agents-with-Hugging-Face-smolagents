```python
# 1. Load .env and confirm your key
import os
from dotenv import load_dotenv

load_dotenv(dotenv_path="/content/tavily.env", override=True)
print("Loaded TAVILY_API_KEY:", os.getenv("TAVILY_API_KEY"))

# 2. Imports and tool definitions
from tavily import TavilyClient
from smolagents import tool, Tool

@tool
def web_search(query: str) -> str:
    """
    Searches the web for your query using the Tavily API.

    Args:
        query: The text to search for.

    Returns:
        A stringified list of search results.
    """
    api_key = os.getenv("TAVILY_API_KEY")
    if not api_key:
        raise ValueError("TAVILY_API_KEY not set in environment")
    client = TavilyClient(api_key=api_key)
    resp = client.search(query)
    return str(resp.get("results", resp))

class VisitWebpageTool(Tool):
    """
    Visits a webpage at the given URL and returns its content as markdown.
    """
    name = "visit_webpage"
    description = (
        "Visits a webpage at the given url and reads its content as markdown."
    )
    inputs = {
        "url": {
            "type": "string",
            "description": "The webpage URL to visit.",
        }
    }
    output_type = "string"

    def forward(self, url: str) -> str:
        """
        Fetches a URL and returns its HTML as markdown.

        Args:
            url: The website address to fetch.

        Returns:
            The page content converted to markdown (truncated if too long).
        """
        import re, requests
        from markdownify import markdownify
        from smolagents.utils import truncate_content

        resp = requests.get(url, timeout=20)
        resp.raise_for_status()
        md = markdownify(resp.text).strip()
        md = re.sub(r"\n{3,}", "\n\n", md)
        return truncate_content(md, max_length=40000)

# 3. Quick test
if __name__ == "__main__":
    query = "What is the melting temperature of vanilla ice cream in °C?"
    print("Query ➞", query)
    print("Results ➞", web_search(query))
```

```python
# agent_example.py

import os
from dotenv import load_dotenv

from tavily import TavilyClient
from smolagents import tool, Tool, CodeAgent, OpenAIServerModel

# 1. Load API key
load_dotenv(dotenv_path="tavily.env", override=True)
API_KEY = os.getenv("TAVILY_API_KEY")
if not API_KEY:
    raise RuntimeError("TAVILY_API_KEY not set in tavily.env")

# 2. Define Tools
@tool
def web_search(query: str) -> str:
    """
    Searches the web for your query using the Tavily API.

    Args:
        query: The text to search for.

    Returns:
        A JSON-stringified list of search results.
    """
    client = TavilyClient(api_key=API_KEY)
    resp = client.search(query)
    return str(resp.get("results", resp))

class VisitWebpageTool(Tool):
    """
    Visits a webpage at the given URL and returns its markdown content.
    """
    name = "visit_webpage"
    description = "Fetch a URL and convert its HTML to markdown."
    inputs = {"url": {"type": "string", "description": "The page URL"}}
    output_type = "string"

    def forward(self, url: str) -> str:
        import re, requests
        from markdownify import markdownify
        from smolagents.utils import truncate_content

        resp = requests.get(url, timeout=20)
        resp.raise_for_status()
        md = markdownify(resp.text).strip()
        md = re.sub(r"\n{3,}", "\n\n", md)
        return truncate_content(md, max_length=40000)

# 3. Create the CodeAgent
model = OpenAIServerModel("gpt-4.1-mini", max_completion_tokens=2000)
agent = CodeAgent(
    model=model,
    tools=[web_search, VisitWebpageTool()],
    max_steps=5,
    name="research_agent",
    description="Uses web search and page visits to answer questions."
)

# 4. Run the agent on a sample query
task = "Find the melting temperature of vanilla ice cream in °C and list the source URL."
print("Agent Task:", task)
result = agent.run(task)

# 5. Display the result
print("\nAgent Final Answer:\n", result)
```

## Create a multi-agent researcher
### Start with web search agent

```python
web_agent = CodeAgent(
    model=OpenAIServerModel(
        #"gpt-4o",
        code_model,  # updated model
        max_completion_tokens=8096,
    ),
    tools=[web_search, VisitWebpageTool()],
    max_steps=10,
    name="web_agent",
    description="Runs web searches for you."
)
web_agent.logger.console.width=66
```

### Setup manager agent

```python
from smolagents import Tool
from typing import Any
from smolagents.utils import make_image_url, encode_image_base64

def check_reasoning_and_plot(final_answer, agent_memory):
    final_answer
    multimodal_model = OpenAIServerModel(
        "gpt-4o",
    )
    filepath = "saved_map.png"
    assert os.path.exists(filepath), "Make sure to save the plot under saved_map.png!"
    image = Image.open(filepath)
    prompt = (
        f"Here is a user-given task and the agent steps: {agent_memory.get_succinct_steps()}. Now here is the plot that was made."
        "Please check that the reasoning process and plot are correct: do they correctly answer the given task?"
        "First list reasons why yes/no, then write your final decision: PASS in caps lock if it is satisfactory, FAIL if it is not."
        "Don't be harsh: if the plot mostly solves the task, it should pass."
        "To pass, a plot should be made using px.scatter_map and not any other method (scatter_map looks nicer)."
        "Also, any run that invents numbers should fail."
    )
    messages = [
        {
            "role": "user",
            "content": [
                {
                    "type": "text",
                    "text": prompt,
                },
                {
                    "type": "image_url",
                    "image_url": {"url": make_image_url(encode_image_base64(image))},
                },
            ],
        }
    ]
    output = multimodal_model(messages).content
    print("Feedback: ", output)
    if "FAIL" in output:
        raise Exception(output)
    return True


manager_agent = CodeAgent(
    model=OpenAIServerModel(
        #"gpt-4o",
        code_model,
        max_tokens=8096,
    ),
    tools=[],
    managed_agents=[web_agent],
    additional_authorized_imports=[
        "geopandas",
        "plotly",
        "plotly.express",  #added
        "plotly.express.colors",  #added
        "shapely",
        "json",
        "pandas",
        "numpy",
    ],
    planning_interval=5,
    verbosity_level=2,
    final_answer_checks=[check_reasoning_and_plot],
    max_steps=15,
)
manager_agent.logger.console.width=66
```
